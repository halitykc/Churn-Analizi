```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Customer-Churn-Records.csv")
data.head()
```

```python
data.iloc[1]
```

```python
data = data.drop(0,axis="index")
data.head()
```

```python
data.info()
```

```python
data.columns
data.describe()
```

```python
data = data.drop(["RowNumber"],axis=1).drop(["CustomerId"],axis=1).drop(["Surname"],axis=1)
data = pd.concat([data.drop(["Gender"],axis=1),pd.get_dummies(data=data["Gender"],dtype=int)],axis=1)
data = pd.concat([data.drop(["Geography"],axis=1),pd.get_dummies(data=data["Geography"],dtype=int)],axis=1)
data = pd.concat([data.drop(["Card Type"],axis=1),pd.get_dummies(data=data["Card Type"],dtype=int)],axis=1)

data.head()


```

```python
import seaborn as sns
corr = data.corr()
import matplotlib.pyplot as plt 

plt.figure(figsize=(20,15))
sns.heatmap(corr, annot=True)
corr["Exited"].sort_values()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

y = data["Exited"].values
X = data.drop(["Exited"],axis=1).values

Xt,Xte,yt,yte =train_test_split(X,y , test_size=0.2,random_state=2)
```

```python
dtc= DecisionTreeClassifier()
dtc.fit(Xt,yt)
```

```python
pred = dtc.predict(Xte)
import sklearn.metrics # type: ignore

print(sklearn.metrics.accuracy_score(yte,pred))
print(sklearn.metrics.classification_report(yte,pred))

sklearn.metrics.confusion_matrix(yte,pred)
```
